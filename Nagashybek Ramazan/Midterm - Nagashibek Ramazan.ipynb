```python
#Task 5. Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить. Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

arr = np.array([2, 2, 2, 3, 3, 3, 5])

def compute(x):
    i = 0
    num = []
    lisT = list(x)

    for n in lisT:
        num.append(lisT.count(n))
        j = i
        if num[i] > 1:
            for n1 in lisT[i + 1:]:
                if n1 == n:
                    del lisT[j]
                    j = j + 1
        i = i + 1
    return lisT, num
result = compute(arr)
    
print("The list of numbers: ", result[0])
print("The number of repetitions of each element: ", result[1])
```

```python
#Task 5. Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить. Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

arr = np.array([2, 2, 2, 3, 3, 3, 5])

def computeV(x):
    lisT, num = np.unique(x, return_counts = True)
    return lisT, num
result = computeV(arr)

print("The list of numbers: ", result[0])
print("The number of repetitions of each element: ", result[1])
```
